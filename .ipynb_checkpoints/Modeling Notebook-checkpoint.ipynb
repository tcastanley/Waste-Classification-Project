```python
train_gen = ImageDataGenerator(rescale=1./255)
val_gen = ImageDataGenerator(rescale=1./255)
test_gen = ImageDataGenerator(rescale=1./255)
```

# Function - Training vs Val Visualization

```python
#Build vizualization function for training/validation loss and accuracy
def visualize_loss_accuracy(results):
    history = results.history
    
    plt.plot(history['loss'], 'b', label='Training Loss')
    plt.plot(history['val_loss'], 'orange', label='Val Loss')
    plt.title('Training vs. Val Loss')
    plt.xlabel('Epochs', fontsize=12)
    plt.ylabel('Losses', fontsize=12)
    plt.legend(loc='best')
    plt.show()
    
    plt.plot(history['accuracy'], 'b', label='Training Acc')
    plt.plot(history['val_accuracy'], 'orange', label='Val Acc')
    plt.title('Training vs. Val Accuracy')
    plt.xlabel('Epochs', fontsize=12)
    plt.ylabel('Accuracy', fontsize=12)
    plt.legend(loc='best')
    plt.show()
```

# Dense Model

## Downsampled - Baseline Dense Model, M=700/70/200

```python
#Bring in the images from the directory.

train_generator = train_gen.flow_from_directory(train_dir,
                                                target_size=(150, 150),
                                                batch_size=700)
val_generator = val_gen.flow_from_directory(val_dir,
                                              target_size=(150, 150),
                                              batch_size=70)
test_generator = val_gen.flow_from_directory(test_dir,
                                              target_size=(150, 150),
                                              batch_size=200)

train_images, train_labels = next(train_generator)
val_images, val_labels = next(val_generator)
test_images, test_labels = next(test_generator)

```

```python
train_images.shape
```

```python
#Reshape the image data for modeling
train_img = train_images.reshape(train_images.shape[0], -1)
val_img = val_images.reshape(val_images.shape[0], -1)
test_img = test_images.reshape(test_images.shape[0], -1)

#Reshape the label data for modeling
train_y = np.reshape(train_labels[:,0], (700, 1))
val_y = np.reshape(val_labels[:,0], (70, 1))
test_y = np.reshape(test_labels[:,0], (200, 1))

print(train_img.shape)
print(train_y.shape)
print(val_img.shape)
print(val_y.shape)
print(test_img.shape)
print(test_y.shape)
```

```python
#Build a baseline model
baseline_model = Sequential()
baseline_model.add(layers.Dense(30, activation='relu', input_shape=(67500,)))
baseline_model.add(layers.Dense(15, activation='relu'))
baseline_model.add(layers.Dense(5, activation='relu'))
baseline_model.add(layers.Dense(1, activation='sigmoid'))

baseline_model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
baseline_results = baseline_model.fit(train_img,
          train_y,
          epochs=50,
          batch_size=10,
          validation_data=(val_img, val_y))
```

```python
visualize_loss_accuracy(baseline_results)
```

```python
baseline_model.summary()
```

```python
vanilla_train_results = baseline_model.evaluate(train_img, train_y)
vanilla_test_results = baseline_model.evaluate(test_img, test_y)

print('Training Results:', vanilla_train_results)
print('Testing Results:', vanilla_test_results)
```

```python
baseline_model.save('Models/baseline_downsampled_model.h5')
```

## Fit_Gen - Baseline CNN Model

```python
cnn_train_generator = train_gen.flow_from_directory(train_dir,
                                                    target_size=(150, 150),
                                                    batch_size=20,
                                                    class_mode='binary')
cnn_val_generator = val_gen.flow_from_directory(val_dir,
                                                target_size=(150, 150),
                                                batch_size=20,
                                                class_mode='binary')
```

```python
#Now lets try out a CNN
base_cnn_model = Sequential()

base_cnn_model.add(layers.Conv2D(32, (3, 3), activation='relu',
                            input_shape=(150, 150, 3)))
base_cnn_model.add(layers.MaxPool2D((2, 2)))

base_cnn_model.add(layers.Conv2D(64, (3, 3), activation='relu'))
base_cnn_model.add(layers.MaxPool2D((2, 2)))
                   
base_cnn_model.add(layers.Conv2D(128, (3, 3), activation='relu'))
base_cnn_model.add(layers.MaxPool2D((2, 2)))
                   
base_cnn_model.add(layers.Conv2D(128, (3, 3), activation='relu'))
base_cnn_model.add(layers.MaxPool2D((2, 2)))
                   
base_cnn_model.add(layers.Flatten())
base_cnn_model.add(layers.Dense(512, activation='relu'))
# base_cnn_model.add(layers.Dense(200, activation='relu'))
# base_cnn_model.add(layers.Dense(20, activation='relu'))
base_cnn_model.add(layers.Dense(1, activation='sigmoid'))

base_cnn_model.compile(optimizer=optimizers.RMSprop(lr=1e-4),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', min_delta=0.005, patience=3, verbose=True)
base_cnn_results = base_cnn_model.fit_generator(cnn_train_generator,
                                                steps_per_epoch=100,
                                                epochs=50,
                                                validation_data=cnn_val_generator)
```

## VGG19

```python
from keras.applications import VGG19
vgg19_model = VGG19(weights='imagenet',
                       include_top=False,
                       input_shape=(150,150,3))
```

```python
vgg19_model.summary()
```

```python
datagen = ImageDataGenerator(rescale=1./255)
batch_size = 20
def extract_features(model, directory, sample_amount):
    features = np.zeros(shape=(sample_amount, 4, 4, 512))
    labels = np.zeros(shape=(sample_amount))
    generator = datagen.flow_from_directory(
        directory, target_size=(150, 150),
        batch_size=20,
        class_mode='binary')
    i=0
    for inputs_batch, labels_batch in generator:
        features_batch = model.predict(inputs_batch)
        features[i * batch_size : (i + 1) * batch_size] = features_batch
        labels[i * batch_size : (i + 1) * batch_size] = labels_batch
        i += 1
        if i * batch_size >= sample_amount:
            break
    return features, labels
```

```python
vgg19_train_features, vgg19_train_labels = extract_features(vgg19_model, train_dir, 700)
vgg19_val_features, vgg19_val_labels = extract_features(vgg19_model, val_dir, 70)
vgg19_test_features, vgg19_test_labels = extract_features(vgg19_model, test_dir, 200)

vgg19_train_features = np.reshape(vgg19_train_features, (700, 4 * 4 * 512))
vgg19_val_features = np.reshape(vgg19_val_features, (70, 4 * 4 * 512))
vgg19_test_features = np.reshape(vgg19_test_features, (200, 4 * 4 * 512))
```

```python
vgg19_train_features.shape
```

```python
#Build a baseline model
vgg19_base_model = Sequential()
vgg19_base_model.add(layers.Dense(200, activation='relu', input_shape=(8192,)))
vgg19_base_model.add(layers.Dense(1, activation='sigmoid'))
vgg19_base_model.compile(optimizer=optimizers.RMSprop(lr=1e-4),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
callbacks = tf.keras.callbacks.EarlyStopping(min_delta=0.05, patience=5)
vgg19_train_base_results = vgg19_base_model.fit(vgg19_train_features,
          vgg19_train_labels,
          epochs=50,
          batch_size=32,
          validation_data=(vgg19_val_features, vgg19_val_labels))

# pre_train_base_results = pre_train_base_model.fit(train_features,
#           pre_train_labels,
#           epochs=50,
#           batch_size=32,
#           validation_data=(val_features, pre_val_labels),
#           callbacks=callbacks)
```

## Xception

```python
from keras.applications import Xception
xcp_model = Xception(weights='imagenet',
                       include_top=False,
                       input_shape=(150,150,3))
```

```python
datagen = ImageDataGenerator(rescale=1./255)
batch_size = 20
def extract_features(model, directory, sample_amount):
    features = np.zeros(shape=(sample_amount, 5, 5, 2048))
    labels = np.zeros(shape=(sample_amount))
    generator = datagen.flow_from_directory(
        directory, target_size=(150, 150),
        batch_size=20,
        class_mode='binary')
    i=0
    for inputs_batch, labels_batch in generator:
        features_batch = model.predict(inputs_batch)
        features[i * batch_size : (i + 1) * batch_size] = features_batch
        labels[i * batch_size : (i + 1) * batch_size] = labels_batch
        i += 1
        if i * batch_size >= sample_amount:
            break
    return features, labels
```

```python
xcp_train_features, xcp_train_labels = extract_features(xcp_model, train_dir, 700)
xcp_val_features, xcp_val_labels = extract_features(xcp_model, val_dir, 70)
xcp_test_features, xcp_test_labels = extract_features(xcp_model, test_dir, 200)

xcp_train_features = np.reshape(xcp_train_features, (700, 5 * 5 * 2048))
xcp_val_features = np.reshape(xcp_val_features, (70, 5 * 5 * 2048))
xcp_test_features = np.reshape(xcp_test_features, (200, 5 * 5 * 2048))
```
